## Dataset - Titanic data
#### Attribute Information
- Survived: Outcome of survival (0 = No; 1 = Yes)
- Pclass: Socio-economic class (1 = Upper class; 2 = Middle class; 3 = Lower class)
- Name: Name of passenger
- Sex: Sex of the passenger
- Age: Age of the passenger (Some entries contain NaN)
- SibSp: Number of siblings and spouses of the passenger aboard
- Parch: Number of parents and children of the passenger aboard
- Ticket: Ticket number of the passenger
- Fare: Fare paid by the passenger
- Cabin: Cabin number of the passenger (Some entries contain NaN)
- Embarked: Port of embarkation of the passenger (C = Cherbourg; Q = Queenstown; S = Southampton)

# 2. Basic plots in Matplotlib library

### 2.1 Bar Plots
- The data for a bar plot is different from the data we've seen up until now. 
- For line charts, we were using numeric data on both the x and y axes. 
- For a bar plot, we need our x-axis data to be a set of labels for the bars, and the y-axis data should be a numeric value that corresponds to the height of each bar.

```python
embarked = titanic_df['Embarked'].value_counts()
plt.bar(x=embarked.index, height=embarked.values,color=['blue', 'green', 'red']);
```

```python
# We can also create a horizontal bar plot using the barh function, but we have to specify the labels on the y-axis the values as bar widths instead of heights.
plt.barh(y=embarked.index, width=embarked.values, color=['blue', 'green', 'red']);
```

### 2.2 Scatter Plots
- To create scatter plots, use Matplotlib's matplotlib.pyplot.scatter function. 
- It expects two lists of values representing x and y coordinates. 
- Let's use the Age and Fare columns from the Titanic data set.

```python
age = titanic_df['Age']
fare = titanic_df['Fare']

plt.scatter(x=age, y=fare);
```

### 2.3 Histograms
- We can use Matplotlib's hist function to display a frequency distribution of some variable. 
- Let's show the distribution of the ages of passengers on the Titanic.

```python
# We can add titles, labels, and gridlines to make it more readable.
ages = titanic_df['Age']
plt.hist(ages)
plt.xlabel('Age')
plt.ylabel('Passengers')
plt.title('Distribution of Passengers Ages')
```

### 2.4 Pie Charts
- To make a pie chart, use Matplotlib's matplotlib.pyplot.pie function, which expects a list representing the sizes of each slice. 
- We can use the count of male and female passengers as an example.

```python
x = titanic_df['Sex'].value_counts()
plt.pie(x, labels=['Male', 'Female'], autopct='%.1f%%'); #we can display the percentage that each slice represents using the autopct parameter by setting it to a format specifier.
```

### 2.5 Box Plots
- You can create box plots (box-and-whisker plots) in Matplotlab using the matplotlib.pyplot.boxplot function. 
- Box plots are important in statistical analysis because they let you see the distrubution of a data set at a glance, including outliers. 
- Let's look at the age of Titanic passengers again, this time in a box plot.

```python
age = titanic_df['Age'].dropna() # Drop missing values before plotting or the boxplot will be empty
fig = plt.figure(figsize=(6, 6))
plt.boxplot(age, labels=['Age']);
```

- We can pick a few key datapoints off the boxplot at a glance. The median age is just under 30, with half of the ages in the 20 to 40 range. The entire distribution of ages is from 0 to 80 years, with a few data points at the high end of the scale considered outliers.

# 3. Basic plots in Seaborn Library

### 3.1.a Distribution Plots - displot()
- Displot helps in understanding the variation in the data distribution. 
- It also provides access to several approaches for visualizing the distribution of data

```python
sns.displot(titanic_df['Age'])
plt.show()
```

- From the above plot we can see that maximum population of the people has an age between 20-38 years

### 3.2.a Categorical Plots - countplot()
- Countplot displays the number of observations in each categorical level using the bars. 
- It is Similar to bar plot

```python
sns.countplot(titanic_df['Embarked'])
plt.show()
```

- It can be clearly observed that majority of the people embarked the titanic ship were from Southampton.

### 3.2.b Categorical Plots - catplot()
- Catplot helps to visualise the categorical data much efficiently. 
- It Can be used for analysis between categorical features and a numerical feature

```python
sns.catplot(x="Survived",y="Fare",data=titanic_df,kind="swarm")
plt.show()
```

- From the above plot we can conclude that there is equal distribution of people who died in the accident irresepective to the ticket fare. 

### 3.3.a Relational Plots - relplot()
- Relplot helps to understand the complicated comparisons much easier. 
- It can be used for analysis between categorical features and a numerical feature

```python
sns.relplot(x="Age", y="Fare", col="Pclass",
            hue="Sex", style="Sex",
            kind="line", data=titanic_df)
plt.show()
```

- It can be observed that most of the women onboard belonged to Pclass=1 or 1st Ticket Class.

### 3.4.a Matrix plots - heatmaps()
- Visualizing data with heatmaps is a great way to do exploratory data analysis, when you have a data set with multiple variables. 
- Heatmaps can reveal general pattern in the dataset, instantly. 
- And it is very easy to make beautiful heatmaps with Seaborn library in Python.

```python
plt.subplots(figsize=(8,5))
sns.heatmap(titanic_df.corr(), cmap = "YlGnBu", annot=True, fmt=".2f")
plt.show()
```

- We observed that the correlation amoung variables are weak.

### 3.5.a MultiPlots grid - facetgrid()
- Creates multiple instances of the same plot on different subsets of the data. Input data must be in a dataframe format
- Facet grid forms a matrix of panels defined by row and column by dividing the variables. 
- Due to these panels, a single plot looks like multiple plots. 
- It is very helpful to analyze all combinations in two discrete variables.

```python
# initialize the FacetGrid object
g = sns.FacetGrid(titanic_df, col='Survived', row='Pclass')
g.map(plt.hist, 'Age')
g.add_legend()
plt.show()
```

- From the above graph it can se observed that a lot of people died from 3rd class ticket and most of them were having age between 20-45

### 3.5.b Multiple grid - pairplot()
- Creates an axis grid and each variable is represented on y-axis and x- axis. 
- Pairwise relationship is represented through scatter plots.
- Unlike FacetGrid, it uses different pair of variable for each subplot.

```python
sns.set_style("ticks")
sns.pairplot(titanic_df, hue='Sex', diag_kind="hist", kind="scatter", palette="husl")
plt.show()
```
