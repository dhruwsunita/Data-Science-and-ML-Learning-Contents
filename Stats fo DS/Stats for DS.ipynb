## **Using Numpy**

```python
data = [2, 5, 3, 8, 6, 9, 4]

sam_var = np.var(data, ddof=1)

pop_var = np.var(data, ddof = 0)

print('sample variance: ', sam_var)
print('population variance:', pop_var)
```

```python
data = [5, 8, 7, 3, 12, 5, 9, 6]
series = pd.Series(data)

sam_var = series.var()
print("varience of sample is ", sam_var)

pop_var = series.var(ddof=0)
print("variance of population", pop_var)
```

```python
data = [5, 9, 8, 3, 6, 7, 2]

sam_var = stats.tvar(data)
print("varience of sample is ", sam_var)

pop_var = stats.tvar(data, ddof=0)
print("variance of population", pop_var)
```

## **Data Representation**

```python
# creating figuers and axes
import matplotlib.pyplot as plt
fig, axs = plt.subplots(3, 2, figsize=(24, 12))
fig.suptitle("Histogram with KDE of various distributions", fontsize=16)

# Normal Destribution(Gaussian Destribution)
data_normal = np.random.normal(loc = 0, scale=1, size=1000)
sns.histplot(data_normal, kde=True, ax = axs[0, 0], color='blue', stat='density', bins=30)
axs[0,0].set_title("Normal Distribution")
axs[0, 0].grid(True)

# Skewed Ditribution
right_skew = np.random.exponential(scale=1, size=1000)
left_skew = -np.random.exponential(scale=1, size=1000)

# Right Skewed Histogram
sns.histplot(right_skew, kde=True, ax=axs[0, 1], color='red', stat='density', bins=30, label='Right Skewed')

# Left Skewed
sns.histplot(left_skew, kde=True, ax=axs[0, 1], color='green', stat='density', bins=30, label='Left Skewed')

axs[0,1].set_title("Skewed Distribution")
axs[0, 1].legend()
axs[0, 1].grid(True)

# Uniform destribution
data_uniform = np.random.uniform(low=0, high=1, size=1000)
sns.histplot(data_uniform, kde=True, ax=axs[1,0], color='purple', stat='density', bins=30)
axs[1,0].set_title('Uniform Destribution')
axs[1,0].grid(True)

# Bimodal Destribution
data_bi = np.concatenate([np.random.normal(loc=2, scale=1, size=500), np.random.normal(loc=5, scale=1, size=500)])
sns.histplot(data_bi, kde=True, ax=axs[1,1], color='orange', stat='density', bins=30)
axs[1,1].set_title("Bimodal Destribution")
axs[1,1].grid(True)

# Multimodal Distribution
data_multi = np.concatenate([np.random.normal(loc=6, scale=1.5, size=500),
                             np.random.normal(loc=0, scale=1, size=500),
                             np.random.normal(loc=6, scale=1.5, size=500)])
sns.histplot(data_multi, kde=True, ax=axs[2, 0], color='brown', stat='density', bins=30)
axs[2, 0].set_title("Multimodal Distribution")
axs[2, 0].grid(True)

# Hide empty suplot
axs[2, 1].axis('off')

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

## **Hypothesis Testing**

**P-value Testing and T-Test**

```python
import numpy as np
from scipy import stats

grp_A = np.random.normal(75, 10, 30) # 75 is mean, 10 is std 30 is sample
grp_B = np.random.normal(80, 12, 30)

# Step 2 perform independent t-test
t_stats, p_value = stats.ttest_ind(grp_A, grp_B)

# Step 3 Display the result
print(f"T-statistics : {t_stats}")
print(f"p-value : {p_value}")

# Step 4 results based on p-value
alpha = 0.05
if p_value <= alpha:
  print("Reject the Null hypothesis")
else:
  print("Fail to reject the Null hypothesis")
```

## **Z-test**

Imagine a company claims that average weight of the product is 500 gram.We wnat to test if this claim is true by collection a sample of 30 products.The sample mean is 505 grams, and the population standard deviation is 10 grams.We will conduct a Z-test to see if the diffrence in the maen is significant or not.

```python
import numpy as np
from scipy import stats


pop_mean = 500
sample_mean = 505
pop_std = 10
sample_size = 30

# Z-value
z_value = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))
print(f"Z-value : {z_value}")

# P-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_value)))
print(f"P-value : {p_value}")

alpha = 0.05

if p_value < alpha:
  print("Reject the null hypothesis")
else:
  print("Fail to reject null hypothesis")
```
