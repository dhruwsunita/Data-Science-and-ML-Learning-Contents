```python
x = df['car_name'][0]  #seperating the company name
x[:x.index(" ")]
```

```python
def com_name(x):
  return x[:x.index(" ")]
```

```python
df['company_name'] = df['car_name'].apply(com_name)
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.head()
```

```python
# seperating car names
def car_name(x):
  return x[x.index(" ")+1:]
```

```python
df['car_name'] = df['car_name'].apply(car_name)
```

```python
# car_price
df['car_prices_in_rupee']
```

```python
df['car_prices_in_rupee'] = df['car_prices_in_rupee'].str.replace(",", "")
```

```python
# currency conversion
def currency_converter(x):
  p = x.split(" ")
  try:
    if p[1] == 'Lakh':
      return str(round(float(p[0])*100000, 1))
    elif p[1] == "Crore":
      return str(round(float(p[0])*10000000, 1))
  except:
    return x
```

```python
df['car_prices_in_rupee'] = df['car_prices_in_rupee'].apply(currency_converter)
```

```python
df.info() # price is still object we will change it to float
```

```python
df.head()
```

```python
df['car_prices_in_rupee'] = df['car_prices_in_rupee'].astype('float64')
```

```python
df['kms_driven']
```

```python
# kms_driven
df['kms_driven'] = df['kms_driven'].str.replace(",", "")
```

```python
# remove kms from number
def remove_kms(x):
  return x.split(" ")[0]
```

```python
df['kms_driven'] = df['kms_driven'].apply(remove_kms)
```

```python
df['kms_driven'] = df['kms_driven'].astype('int64')
```

```python
df.head()
```

```python
# engine
df['engine']
```

```python
df['engine'] = df['engine'].apply(lambda x:x.split(" ")[0])
```

```python
df['engine'] = df['engine'].astype("int64")
```

```python
# Seats
df['Seats']
```

```python
df['Seats'] = df['Seats'].apply(lambda x:x.split(" ")[0])
```

```python
df['Seats'] = df['Seats'].astype("int64")
```

```python
# manufacture date
df['manufacture']
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
# car_name
le = LabelEncoder()
df['car_name'] = le.fit_transform(df['car_name'])
```

```python
# fuel_type
df['fuel_type'] = le.fit_transform(df['fuel_type'])
```

```python
# transmission
df['transmission'] = le.fit_transform(df['transmission'])
```

```python
# outlier detection
# plt.figure(figsize=(5,6))
# sns.boxplot(x="car_name", data = df)
# plt.show()


# List of column names
columns = ['car_name', 'car_prices_in_rupee', 'kms_driven', 'fuel_type',
           'transmission', 'ownership', 'manufacture', 'engine', 'Seats',
           'company_name']

# Number of plots
num_cols = len(columns)

# Create subplots
fig, axes = plt.subplots(nrows=(num_cols // 3) + 1, ncols=3, figsize=(15, 10))

# Flatten axes array for easy iteration
axes = axes.flatten()

# Plot boxplots for each column
for i, col in enumerate(columns):
    if col in df.select_dtypes(include=['int64', 'float64']).columns:
        sns.boxplot(data=df, y=col, ax=axes[i])
        axes[i].set_title(f'Boxplot of {col}')
    elif col in df.select_dtypes(include=['object', 'category']).columns:
        sns.boxplot(data=df, x=col, ax=axes[i])
        axes[i].set_title(f'Boxplot of {col}')
    else:
        axes[i].axis('off')  # Skip plotting for incompatible columns

# Hide any remaining empty subplots
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# removing outlier of engine using IQR method
Q1 = np.quantile(df['engine'], 0.25)
Q3 = np.quantile(df['engine'], 0.75)
```

```python
IQR = Q3 - Q1
```

```python
min_r = Q1 - (1.5*IQR)
max_r = Q3 + (1.5*IQR)
```

```python
min_r, max_r
```

```python
df = df[df['engine'] <= max_r]
```

```python
# scaling the data
scaler = StandardScaler()
new_df = pd.DataFrame(scaler.fit_transform(df), columns = df.columns)
```
